```python
import os
import numpy as np
import pandas as pd
import skopt
```

```python
# this is old cv train result
cv_train_result_root_dir = '/zfssz2/ST_MCHRI/BIGDATA/PROJECT/NIPT_CNV/f_cnv_out/model111/model_csvlogger'

train_csv_fn = 'b512_ep100_lr0.001_dr0.5_fc64_blk443_win1000_l2r0.0001_temp6_lblsmt0_pw-8-cnvnet'


train_result_csv_fname=os.path.join(cv_train_result_root_dir, train_csv_fn)

train_df = pd.read_csv(train_result_csv_fname, sep=',')

train_df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot( 'epoch', 'val_loss', data=train_df, marker='.', 
         linestyle='-', color='r', linewidth=1.5, label='valiation loss')
plt.plot( 'epoch', 'loss', data=train_df, marker='*',
         color='b', linewidth=1.5, label='train loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend()
```

```python
plt.plot( 'epoch', 'val_acc', data=train_df, marker='.', 
         linestyle='-', color='r', linewidth=1.5, label='valiation accuracy')
plt.plot( 'epoch', 'acc', data=train_df, marker='*',
         color='b', linewidth=1.5, label='train accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.grid(True)
plt.legend()
```
